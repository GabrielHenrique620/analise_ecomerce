```python
import kagglehub
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np
```

```python

# Download do arquivo utilizado para an√°lise
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

# Comando utilizado para localizar o path onde o arquivo foi baixado
print("Path to dataset files:", path)

# O path retornado √© o diret√≥rio base, formata√ß√£o do path retornado para leitura do DataFrame
base_path = r"C:/Users/Gabriel Henrique/.cache/kagglehub\datasets/olistbr/brazilian-ecommerce/versions/2"

# Lista os arquivos no diret√≥rio
files = os.listdir(base_path)
print("Arquivos dispon√≠veis:", files)

# Variaveis para defini√ß√£o de cada arquivo utilizado

# DataFrame com dados relacionados aos pedidos:
orders_df = pd.read_csv(os.path.join(base_path, "olist_orders_dataset.csv"))

# DataFrame com dados relacionados aos produtos.
products_df = pd.read_csv(os.path.join(base_path, "olist_products_dataset.csv"))

# DataFrame com dados relacionados aos clientes.
customers_df = pd.read_csv(os.path.join(base_path, "olist_customers_dataset.csv"))

# Iniciando a an√°lise dos dataframes
display(orders_df.head(),products_df.head(),customers_df.head())
```

```python

# Configura√ß√µes de visualiza√ß√£o
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 6]

# An√°lise temporal de vendas
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])

# Criando an√°lise di√°ria
daily_orders = orders_df.resample('D', on='order_purchase_timestamp').size()

# Criando m√©dia m√≥vel de 7 dias
moving_average = daily_orders.rolling(window=7).mean()

# Calculando estat√≠sticas do per√≠odo espec√≠fico
periodo_interesse = daily_orders['2017-10':'2018-01']
media_periodo = periodo_interesse.mean()
max_periodo = periodo_interesse.max()
crescimento = ((periodo_interesse[-1] / periodo_interesse[0]) - 1) * 100

# Plotando o gr√°fico
plt.figure(figsize=(15, 7))

# Plotando os dados completos
plt.plot(daily_orders.index, daily_orders.values, label='Vendas Di√°rias', alpha=0.5, color='gray')
plt.plot(moving_average.index, moving_average.values, label='M√©dia M√≥vel (7 dias)', 
         linewidth=2, color='blue')

# Destacando o per√≠odo de interesse
plt.axvspan('2017-10-01', '2018-01-31', alpha=0.2, color='yellow', 
            label='Per√≠odo de Crescimento')

# Melhorando a apar√™ncia
plt.title('Tend√™ncia de Vendas Di√°rias com M√©dia M√≥vel', pad=20, fontsize=14)
plt.xlabel('Data', fontsize=12)
plt.ylabel('N√∫mero de Pedidos', fontsize=12)
plt.legend()
plt.xticks(rotation=45)

# Adicionando informa√ß√µes sobre o per√≠odo destacado
info_text = (f"Per√≠odo Out/2017 - Jan/2018:\n"
            f"M√©dia di√°ria: {media_periodo:.0f} pedidos\n"
            f"Pico: {max_periodo:.0f} pedidos\n"
            f"Crescimento no per√≠odo: {crescimento:.1f}%")

plt.text(0.02, 0.98, info_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

# Imprimindo an√°lise detalhada
print("\nAn√°lise do Per√≠odo de Crescimento (Out/2017 - Jan/2018):")
print("-" * 50)
print(f"M√©dia di√°ria de pedidos: {media_periodo:.0f}")
print(f"Pico de pedidos: {max_periodo:.0f}")
print(f"Crescimento no per√≠odo: {crescimento:.1f}%")

# An√°lise mensal do per√≠odo
monthly_orders = daily_orders.resample('M').sum()
print("\nVendas mensais no per√≠odo:")
for date, value in monthly_orders['2017-10':'2018-01'].items():
    print(f"{date.strftime('%B/%Y')}: {value:,.0f} pedidos")
```

```python
plt.figure(figsize=(16, 7))

status_counts = orders_df['order_status'].value_counts()
total_orders = len(orders_df)
status_pcts = (status_counts / total_orders * 100).round(1)

delivered_orders = status_counts.get('delivered', 0)
non_delivered = total_orders - delivered_orders
delivery_rate = (delivered_orders / total_orders * 100).round(1)
non_delivery_rate = (non_delivered / total_orders * 100).round(1)


ax = status_counts.plot(kind='bar', color='skyblue')

# Ajustando a posi√ß√£o dos textos acima das barras
for i, (count, pct) in enumerate(zip(status_counts, status_pcts)):
    # Calculando um deslocamento vertical baseado no valor m√°ximo
    y_offset = status_counts.max() * 0.02  # 2% do valor m√°ximo
    
    # Posicionando o texto um pouco acima do topo da barra
    ax.text(i, count + y_offset, f'{count:,}\n({pct}%)', 
            ha='center', va='bottom')

plt.title('Distribui√ß√£o de Status dos Pedidos', pad=20, fontsize=14)
plt.xlabel('Status do Pedido', fontsize=12)
plt.ylabel('Quantidade de Pedidos')

plt.xticks(rotation=45, ha='right')

# Ajustando os limites do eixo y para acomodar os textos
plt.ylim(0, status_counts.max() * 1.15)  # Aumentando o limite superior em 15%

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

info_text = (f"Total de Pedidos: {total_orders:,}\n"
            f"Pedidos Entregues: {delivered_orders:,} ({delivery_rate}%)\n"
            f"Pedidos N√£o Entregues: {non_delivered:,} ({non_delivery_rate}%)\n"
            )

plt.text(0.95, 0.95, info_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
# An√°lise por dia da semana
plt.figure(figsize=(12, 7))

# Preparando os dados
orders_df['weekday'] = orders_df['order_purchase_timestamp'].dt.day_name()
weekday_order = orders_df['weekday'].value_counts()

# Reordenando os dias da semana
week_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_order = weekday_order.reindex(week_order)

# Calculando percentuais
total_orders = weekday_order.sum()
weekday_pcts = (weekday_order / total_orders * 100).round(1)

# Criando o gr√°fico
ax = weekday_order.plot(kind='bar', color='skyblue')

# Adicionando valores sobre as barras
for i, (count, pct) in enumerate(zip(weekday_order, weekday_pcts)):
    y_offset = weekday_order.max() * 0.02
    ax.text(i, count + y_offset, f'{count:,}\n({pct}%)', 
            ha='center', va='bottom')

# Melhorando a apar√™ncia
plt.title('Distribui√ß√£o de Pedidos por Dia da Semana', pad=20, fontsize=14)
plt.xlabel('Dia da Semana', fontsize=12)
plt.ylabel('N√∫mero de Pedidos', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Ajustando limites do eixo y
plt.ylim(0, weekday_order.max() * 1.15)

# Adicionando grid
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Adicionando box com informa√ß√µes
weekday_max = weekday_order.index[weekday_order.argmax()]
weekday_min = weekday_order.index[weekday_order.argmin()]

info_text = (f"Total de Pedidos: {total_orders:,}\n"
            f"Dia mais movimentado: {weekday_max}\n"
            f"({weekday_order[weekday_max]:,} pedidos, {weekday_pcts[weekday_max]}%)\n"
            f"Dia menos movimentado: {weekday_min}\n"
            f"({weekday_order[weekday_min]:,} pedidos, {weekday_pcts[weekday_min]}%)")

plt.text(0.95, 0.95, info_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()


```

```python
# An√°lise de pedidos por m√™s
plt.figure(figsize=(14, 7))

# Preparando os dados
orders_df['month'] = orders_df['order_purchase_timestamp'].dt.month
# Convertendo n√∫meros dos meses para nomes
month_names = {
    1: 'Janeiro', 2: 'Fevereiro', 3: 'Mar√ßo', 4: 'Abril',
    5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
    9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
}
orders_df['month_name'] = orders_df['month'].map(month_names)
monthly_orders = orders_df['month'].value_counts().sort_index()
monthly_pcts = (monthly_orders / len(orders_df) * 100).round(1)

# Criando o gr√°fico
ax = monthly_orders.plot(kind='bar', color='skyblue')

# Adicionando valores sobre as barras
for i, (count, pct) in enumerate(zip(monthly_orders, monthly_pcts)):
    y_offset = monthly_orders.max() * 0.02
    ax.text(i, count + y_offset, f'{count:,}\n({pct}%)', 
            ha='center', va='bottom')

# Melhorando a apar√™ncia
plt.title('Distribui√ß√£o de Pedidos por M√™s', pad=20, fontsize=14)
plt.xlabel('M√™s', fontsize=12)
plt.ylabel('N√∫mero de Pedidos', fontsize=12)

# Ajustando os labels do eixo x para nomes dos meses
plt.xticks(range(12), [month_names[i] for i in range(1, 13)], rotation=45, ha='right')

# Ajustando limites do eixo y
plt.ylim(0, monthly_orders.max() * 1.15)

# Adicionando grid
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Calculando estat√≠sticas
peak_month = monthly_orders.idxmax()
low_month = monthly_orders.idxmin()
avg_monthly = monthly_orders.mean()

# Adicionando box com informa√ß√µes
info_text = (f"Total de Pedidos: {len(orders_df):,}\n"
            f"M√©dia Mensal: {avg_monthly:.0f} pedidos\n"
            f"M√™s Mais Movimentado: {month_names[peak_month]}\n"
            f"({monthly_orders[peak_month]:,} pedidos, {monthly_pcts[peak_month]}%)\n"
            f"M√™s Menos Movimentado: {month_names[low_month]}\n"
            f"({monthly_orders[low_month]:,} pedidos, {monthly_pcts[low_month]}%)")

plt.text(0.95, 0.95, info_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

# Imprimindo an√°lise detalhada
print("\nAn√°lise Detalhada por M√™s:")
print("-" * 50)
print(f"M√©dia mensal de pedidos: {avg_monthly:.0f}")
print(f"Varia√ß√£o sazonal (Max/Min): {(monthly_orders.max()/monthly_orders.min() - 1)*100:.1f}%")
print("\nDistribui√ß√£o mensal:")
for month in range(1, 13):
    print(f"{month_names[month]}: {monthly_orders.get(month, 0):,} pedidos ({monthly_pcts.get(month, 0):.1f}%)")
```

```python
# An√°lise de tempo de entrega
orders_df['delivery_time'] = (pd.to_datetime(orders_df['order_delivered_customer_date']) - 
                            pd.to_datetime(orders_df['order_purchase_timestamp'])).dt.days

plt.figure(figsize=(15, 8))

# Criando boxplot com ajustes
sns.boxplot(data=orders_df, y='delivery_time', color='skyblue', showfliers=True)

# Calculando estat√≠sticas importantes
median_delivery = orders_df['delivery_time'].median()
mean_delivery = orders_df['delivery_time'].mean()
std_delivery = orders_df['delivery_time'].std()
on_time_deliveries = (orders_df['delivery_time'] <= 15).mean() * 100  # Considerando 15 dias como prazo padr√£o

# Ajustando os limites do eixo y
y_min = orders_df['delivery_time'].quantile(0.01)
y_max = orders_df['delivery_time'].quantile(0.99)
plt.ylim(y_min, y_max)

# Adicionando grid
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# T√≠tulo e labels
plt.title('An√°lise do Tempo de Entrega', pad=20, fontsize=14)
plt.ylabel('Dias para Entrega', fontsize=12)

# Criando texto com estat√≠sticas mais detalhadas
stats_text = (
    f"Estat√≠sticas de Entrega:\n"
    f"Mediana: {median_delivery:.1f} dias\n"
    f"M√©dia: {mean_delivery:.1f} dias\n"
    f"Desvio Padr√£o: {std_delivery:.1f} dias\n"
    f"Entregas em at√© 15 dias: {on_time_deliveries:.1f}%\n"
    f"90% das entregas: at√© {orders_df['delivery_time'].quantile(0.9):.1f} dias\n"
    f"95% das entregas: at√© {orders_df['delivery_time'].quantile(0.95):.1f} dias"
)

# Adicionando box com estat√≠sticas
plt.text(0.02, 0.98, stats_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Adicionando linhas de refer√™ncia para prazos comuns
plt.axhline(y=7, color='green', linestyle='--', alpha=0.5, label='Meta 7 dias')
plt.axhline(y=15, color='orange', linestyle='--', alpha=0.5, label='Meta 15 dias')
plt.legend()

plt.tight_layout()
plt.show()

# An√°lise detalhada dos percentis
print("\nAn√°lise Detalhada dos Tempos de Entrega:")
print("-" * 50)

# Estat√≠sticas b√°sicas
print(f"\nEstat√≠sticas Gerais:")
print(f"Total de entregas analisadas: {len(orders_df):,}")
print(f"Tempo m√©dio de entrega: {mean_delivery:.1f} dias")
print(f"Desvio padr√£o: {std_delivery:.1f} dias")
print(f"Mediana do tempo de entrega: {median_delivery:.1f} dias")

# An√°lise por faixas de tempo
print("\nDistribui√ß√£o por Faixas de Tempo:")
faixas = [
    (0, 3, "At√© 3 dias"),
    (4, 7, "4-7 dias"),
    (8, 15, "8-15 dias"),
    (16, 30, "16-30 dias"),
    (31, float('inf'), "Mais de 30 dias")
]

for min_dias, max_dias, label in faixas:
    n_entregas = len(orders_df[(orders_df['delivery_time'] >= min_dias) & 
                              (orders_df['delivery_time'] <= max_dias)])
    percentual = (n_entregas / len(orders_df)) * 100
    print(f"{label}: {n_entregas:,} entregas ({percentual:.1f}%)")

# Percentis detalhados
print("\nPercentis do Tempo de Entrega:")
percentis = [10, 25, 50, 75, 90, 95, 99]
for p in percentis:
    valor = orders_df['delivery_time'].quantile(p/100)
    print(f"{p}¬∫ percentil: {valor:.1f} dias")
```

```python
# Criando as colunas relacionadas a diferen√ßa entre os dias
orders_df['real_delivery_time'] = (pd.to_datetime(orders_df['order_delivered_customer_date']) - 
                                 pd.to_datetime(orders_df['order_purchase_timestamp'])).dt.days

orders_df['estimated_delivery_time'] = (pd.to_datetime(orders_df['order_estimated_delivery_date']) - 
                                      pd.to_datetime(orders_df['order_purchase_timestamp'])).dt.days

# Criando a coluna on_time
orders_df['on_time'] = orders_df['real_delivery_time'] <= orders_df['estimated_delivery_time']

# Agora fazemos o merge
delivery_analysis = pd.merge(
    orders_df[['order_id', 'real_delivery_time', 'estimated_delivery_time', 'on_time', 'customer_id']],
    customers_df[['customer_id', 'customer_city']],
    on='customer_id'
)

# O resto do c√≥digo continua igual...

# Calculando m√©tricas por cidade
city_metrics = delivery_analysis.groupby('customer_city').agg({
   'real_delivery_time': 'mean',
   'estimated_delivery_time': 'mean',
   'on_time': 'mean',
   'order_id': 'count'
}).reset_index()

# Renomeando colunas
city_metrics.columns = ['city', 'avg_delivery_time', 'avg_estimated_time', 'on_time_rate', 'total_orders']
city_metrics['on_time_rate'] = city_metrics['on_time_rate'] * 100

# Filtrando top 15 cidades por volume de pedidos
top_cities = city_metrics.nlargest(15, 'total_orders')

# Primeiro gr√°fico: Tempo m√©dio real vs estimado por cidade
plt.figure(figsize=(15, 8))
x = np.arange(len(top_cities))
width = 0.35

plt.bar(x - width/2, top_cities['avg_delivery_time'], width, label='Tempo Real', color='skyblue')
plt.bar(x + width/2, top_cities['avg_estimated_time'], width, label='Tempo Estimado', color='lightgreen')

plt.title('Tempo de Entrega vs Estimado por Cidade (Top 15 em Volume)', pad=20, fontsize=14)
plt.xlabel('Cidade', fontsize=12)
plt.ylabel('Dias', fontsize=12)
plt.xticks(x, top_cities['city'], rotation=45, ha='right')
plt.legend()

# Adicionando os valores sobre as barras com offset ajustado
for i, (real, est) in enumerate(zip(top_cities['avg_delivery_time'], top_cities['avg_estimated_time'])):
   plt.text(i - width/2, real + 0.5, f'{real:.1f}', ha='center', va='bottom')
   plt.text(i + width/2, est + 0.5, f'{est:.1f}', ha='center', va='bottom')

# Ajustando limites do eixo y para dar mais espa√ßo
plt.ylim(0, max(max(top_cities['avg_delivery_time']), max(top_cities['avg_estimated_time'])) * 1.15)

plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Segundo gr√°fico: Taxa de entregas no prazo por cidade
plt.figure(figsize=(15, 8))
bars = plt.bar(top_cities['city'], top_cities['on_time_rate'], color='skyblue')
plt.title('Taxa de Entregas no Prazo por Cidade (Top 15 em Volume)', pad=20, fontsize=14)
plt.xlabel('Cidade', fontsize=12)
plt.ylabel('% de Entregas no Prazo', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Adicionando os valores sobre as barras com offset ajustado
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.3,
             f'{height:.1f}%',
             ha='center', va='bottom')

# Ajustando limites do eixo y para dar mais espa√ßo
y_max = max(top_cities['on_time_rate']) * 1.21  # 21% de espa√ßo adicional
plt.ylim(0, y_max)

# Adicionando linha m√©dia
media_geral = top_cities['on_time_rate'].mean()
plt.axhline(y=media_geral, color='red', linestyle='--', 
            label=f'M√©dia Geral: {media_geral:.1f}%')

# Texto da m√©dia no topo do gr√°fico (dentro da √°rea)
plt.text(0.5, 0.98, f'M√©dia Geral: {media_geral:.1f}%',
         transform=plt.gca().transAxes,
         horizontalalignment='center',
         verticalalignment='top',
         bbox=dict(facecolor='white', alpha=0.9, edgecolor='red', linewidth=1))

# Texto informativo no canto superior esquerdo (dentro da √°rea)
info_text = (
    f"Estat√≠sticas:\n"
    f"Melhor performance: {top_cities.iloc[top_cities['on_time_rate'].argmax()]['city']}"
    f" ({top_cities['on_time_rate'].max():.1f}%)\n"
    f"Pior performance: {top_cities.iloc[top_cities['on_time_rate'].argmin()]['city']}"
    f" ({top_cities['on_time_rate'].min():.1f}%)"
)

plt.text(0.02, 0.98, info_text,
         transform=plt.gca().transAxes,
         verticalalignment='top',
         horizontalalignment='left',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.9))

plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
